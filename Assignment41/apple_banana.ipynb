```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
apple_w = np.random.normal(6,.5,100)
apple_h = np.random.normal(6,.5,100)

banana_w = np.random.normal(4,.5,100)
banana_h = np.random.normal(8,.5,100)

strawberry_w = np.random.normal(3,.5,100)
strawberry_l = np.random.normal(4,.5,100)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(apple_w,apple_h,c='r',label='Apple')
plt.scatter(banana_w,banana_h,c='y',label='Banana')
plt.scatter(strawberry_w,strawberry_l,c='purple',label='Strawberry')
plt.xlabel('height')
plt.ylabel('width')
plt.title('Apple,Banana,Strawberry')
plt.legend()
```

```python
# KNN: K nearest neighbore

width = np.concatenate((apple_w,banana_w,strawberry_w))
length = np.concatenate((apple_h,banana_h,strawberry_l))
X_train = np.array([width,length]).T
X_train.shape
```

```python
Y_apples = np.zeros(n,dtype='int')
Y_bananas = np.ones(n,dtype='int')
Y_strawberry = np.ones(n,dtype='int')*2
```

```python
Y_train = np.concatenate((Y_apples,Y_bananas,Y_strawberry))
Y_train
```

```python
def euc_distance(x1,x2):
    return np.sqrt(np.sum((x1 - x2)**2))
```

```python
new_fruit = np.array([15,15])

distances = []

for x in X_train:
    d = euc_distance(new_fruit,x)
    distances.append(d)

distances
```

```python
k = 9
```

```python
nearest_neighbores = np.argsort(distances)[0:k]
nearest_neighbores
```

```python
Y_train[nearest_neighbores]
```

```python
res = np.bincount(Y_train[nearest_neighbores])
```

```python
if np.argmax(res) == 0:
    print("Apple")
if np.argmax(res) == 1:
    print("Banana")
if np.argmax(res) == 2:
    print("Strawberry")
```
