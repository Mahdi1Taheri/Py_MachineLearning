```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.neighbors import KNeighborsClassifier
```

```python


def gen_data(n):
    apple_w = np.random.normal(6,1,n)
    apple_h = np.random.normal(6,1,n)
    banana_w = np.random.normal(4,1,n)
    banana_h = np.random.normal(8,1,n)

    width = np.concatenate((apple_w,banana_w))
    length = np.concatenate((apple_h,banana_h))
    X = np.array([width,length]).T

    Y_apples = np.zeros(n,dtype='int')
    Y_bananas = np.ones(n,dtype='int')
    Y = np.concatenate((Y_apples,Y_bananas))

    return X,Y
```
