```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from perceptron import Perceptron
```

```python
df = pd.read_csv('input\Surgical-deepnet.csv')
df.head()
```

```python
X = df.drop('complication', axis=1).copy().values
Y = df['complication'].copy().values

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=.2)
# Y_test = Y_test.reshape(-1,1)
epochs = 50
X_test.shape, Y_test.shape
```

```python
# lr sigmoid: .0001
# lr relu: .0000001
model = Perceptron(.00001,X_train.shape[1])
model.fit(X_train, Y_train,X_test, Y_test,epochs,function='sigmoid')

```

```python
model.calc_accuracy(X_test, Y_test)
```
